```python
# Combine relevant columns for content-based features
df_new['combined_features'] = (df_new['cuisines'] + ' ' + df_new['name'] + ' ' + df_new['address'] + ' ' + df_new['city']).str.lower()

```

```python
# Initialize TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD


tfidf = TfidfVectorizer(stop_words='english')
```

```python
# Create the TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(df_new['combined_features'])
```

```python
from sklearn.neighbors import NearestNeighbors

# Fit NearestNeighbors model
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(tfidf_matrix)

# Find top 10 neighbors for each row
distances, indices = knn.kneighbors(tfidf_matrix, n_neighbors=10)
```

```python
# Clean and normalize the 'cuisines' column (lowercase, remove extra spaces)
df_new['cuisines'] = df_new['cuisines'].str.lower().str.replace(r'\s+', ' ', regex=True).str.strip()

def recommend_restaurants_knn(restaurant_id, num_recommendations=10, city=None, min_cost=None, max_cost=None, cuisine=None):
    try:
        # Get the index of the given restaurant
        idx = df_new[df_new['res_id'] == restaurant_id].index[0]

        # Get the indices of the nearest neighbors, excluding the restaurant itself
        neighbor_indices = indices[idx][1:]  # Exclude the first index (self)

        # Create a dataframe with the neighbor restaurants
        neighbors_df = df_new.iloc[neighbor_indices].copy()

        # Apply city filter
        if city:
            neighbors_df = neighbors_df[neighbors_df['city'].str.lower() == city.lower()]
            print("After city filter:", neighbors_df.shape)

        # Apply cost filters
        if min_cost is not None:
            neighbors_df = neighbors_df[neighbors_df['average_cost_for_two'] >= min_cost]
            print("After min cost filter:", neighbors_df.shape)

        if max_cost is not None:
            neighbors_df = neighbors_df[neighbors_df['average_cost_for_two'] <= max_cost]
            print("After max cost filter:", neighbors_df.shape)

        # Apply cuisine filter
        if cuisine:
            # Normalize the cuisine input to a list
            cuisine_list = [c.strip().lower() for c in cuisine.split(",")] if isinstance(cuisine, str) else cuisine
            neighbors_df = neighbors_df[
                neighbors_df['cuisines'].apply(lambda x: any(c in x for c in cuisine_list) if isinstance(x, str) else False)
            ]
            print("After cuisine filter:", neighbors_df.shape)

        # Check if any restaurants remain after filtering
        if neighbors_df.empty:
            print("No recommendations found after applying filters.")
            return None

        # Select top N recommended restaurants
        neighbors_df = neighbors_df.head(num_recommendations)

        # Return the recommended restaurants' details
        return neighbors_df[['name', 'address', 'cuisines', 'average_cost_for_two', 'aggregate_rating']]

    except IndexError:
        print(f"Restaurant ID {restaurant_id} not found in the dataset.")
        return None
    except Exception as e:
        print("An error occurred:", str(e))
        return None

# Example test
recommended = recommend_restaurants_knn(
    restaurant_id=3400299,  # Replace with the valid restaurant ID
    num_recommendations=10,
    city="Agra",  # Apply city filter
    min_cost=500,  # Apply less restrictive cost filter
    max_cost=5000,  # Apply less restrictive cost filter
    cuisine="chinese"  # Filter for multiple cuisines
)

print("Recommended Restaurants:")
print(recommended)
```
